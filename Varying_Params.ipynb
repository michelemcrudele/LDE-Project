```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
import networkx as nx
import sys
from utils import SIR_net, SIR_net_adaptive
```

```python
N = 1000 # number of nodes

ave_degree = 15

initial_infecteds = np.random.choice(np.arange(N), size=100, replace=False)

initial_no_vax = np.random.choice(np.arange(N), size=int(12/100*N), replace=False) #in italy novax = 12% population

BAG1 = nx.barabasi_albert_graph(N, ave_degree//2)

pro = 0.3        # rate of classical media influence on people
pol = 0.02       # propensity of opinion polarization
p_sym = 0.05     # prob for an infected person to have severe symptoms
r = 0.8         # rate of vaccination for PV
```

```python
R0s = np.linspace(1.2, 3.5,  12) # sensible COVID-19 range
# beta = R0*mu
```

# SIMULATIONS

###Â Here I fix mu and vary R0

```python
mu = 0.1    # recovery rate

fig, ax = plt.subplots(nrows=3, ncols=4, figsize=(4*4, 3*3))
fig.tight_layout(h_pad = 4)
i = 0

for R0 in R0s:
    BAG2 = BAG1.copy()
    BAG3 = BAG2.copy()
    print('1 simulation:\n')
    sim1 = SIR_net(BAG1, beta = mu*R0, mu = mu, initial_infecteds = initial_infecteds)
    
    print('\n2 simulation:\n')
    sim2 = SIR_net_adaptive(BAG1, BAG2, beta = mu*R0, mu = mu, r = r, pro = pro, pol = pol, p_sym = p_sym, 
                            initial_infecteds = initial_infecteds, initial_no_vax=[])
    
    print('\n3 simulation:\n')
    sim3 = SIR_net_adaptive(BAG1, BAG3, beta = mu*R0, mu = mu, r = r, pro = pro, pol = pol, p_sym = p_sym, 
                            initial_infecteds = initial_infecteds, initial_no_vax = initial_no_vax)
    
    colors = cm.get_cmap('winter', 3)
    
    if i<=3:
        ax[0, i].plot(sim2[0], sim2[2], lw=3, c=colors(0), label= "0 No-Vax", alpha=0.6)
        ax[0, i].plot(sim3[0], sim3[2], lw=3, c=colors(1), label= "12% No-Vax", alpha=0.6)
        ax[0, i].plot(sim1[0], sim1[2], lw=3, c=colors(4), label= "SIR", alpha=0.6)
        ax[0, i].legend(fontsize=10)
        ax[0, i].set_title("R0 = %.2f"%R0)
        ax[0, i].set_ylabel("Prevalence", size=8)
        ax[0, i].set_xlabel("time", size=8)
        ax[0, i].set_xlim(0, 100)
        ax[0, i].set_ylim(0, 1000)

    elif i<=7:
        ax[1, i-4].plot(sim2[0], sim2[2], lw=3, c=colors(0), label= "0 No-Vax", alpha=0.6)
        ax[1, i-4].plot(sim3[0], sim3[2], lw=3, c=colors(1), label= "12% No-Vax", alpha=0.6)
        ax[1, i-4].plot(sim1[0], sim1[2], lw=3, c=colors(4), label= "SIR", alpha=0.6)
        ax[1, i-4].legend(fontsize=10)
        ax[1, i-4].set_title("R0 = %.2f"%R0)
        ax[1, i-4].set_ylabel("Prevalence", size=8)
        ax[1, i-4].set_xlabel("time", size=8)
        ax[1, i-4].set_xlim(0, 100)
        ax[1, i-4].set_ylim(0, 1000)
    
    else:
        ax[2, i-8].plot(sim2[0], sim2[2], lw=3, c=colors(0), label= "0 No-Vax", alpha=0.6)
        ax[2, i-8].plot(sim3[0], sim3[2], lw=3, c=colors(1), label= "12% No-Vax", alpha=0.6)
        ax[2, i-8].plot(sim1[0], sim1[2], lw=3, c=colors(4), label= "SIR", alpha=0.6)
        ax[2, i-8].legend(fontsize=10)
        ax[2, i-8].set_title("R0 = %.2f"%R0)
        ax[2, i-8].set_ylabel("Prevalence", size=8)
        ax[2, i-8].set_xlabel("time", size=8)
        ax[2, i-8].set_xlim(0, 100)
        ax[2, i-8].set_ylim(0, 1000)
    
    i += 1  
    
plt.show()
```
